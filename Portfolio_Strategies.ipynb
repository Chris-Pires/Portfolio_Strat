```python
#On commence par paramétrer nos plots
fontsize = 12  
width_in_cm = 28.  


width_in_inches = width_in_cm * 0.3937
height_in_inches = width_in_inches * 9. / 16. 
params = {
    'figure.figsize': (width_in_inches, height_in_inches),  
    'axes.labelsize': fontsize,
    'axes.titlesize': fontsize,
    'xtick.labelsize': fontsize * 1.2,  
    'ytick.labelsize': fontsize * 1.2,
    'legend.fontsize': fontsize * 1.2,
    'lines.linewidth': 2,  
    'lines.markersize': 6,  
}

plt.rcParams.update(params)
```

```python
#On calcule les returns de nos 10 stocks sur l'année 2019 :
returns = agg_price_data_2019_10.pct_change().dropna()
```

```python
# On définit une classe 'Strategy' qui va être utilisée pour les trois stratégies de portfolio, et qui contient des fonctions qui vont être utiles à notre étude, comme le maximum drawdown et le sharpe ratio
class Strategy:
    def __init__(self, name, returns, prices):
        self.name = name
        self.returns = returns
        self.prices = prices
        self.pnl = np.c_[np.array([1000000.]), 1000000.*np.cumprod(1.+returns.to_numpy()).reshape(1,-1)].flatten()


    def show(self):
        return pd.DataFrame(self.returns)

    def show_covariance_matrix(self):
        return sn.heatmap(self.prices.corr(),annot=True,cmap='OrRd')

    def volatility(self):
        return np.std(self.returns)*np.sqrt(252)

    def sharpe_ratio(self): # Fonction de calcul du sharpe ratio
        return np.mean(self.returns)*np.sqrt(252)/np.std(self.returns)

    def max_dd(self): # Maximum drawdown
        return 1. - np.min(np.flip(np.minimum.accumulate(np.flip(self.pnl)))/self.pnl)

    def illustrate(self): # Fonction pour paramétrer les plots
        fig, ax = plt.subplots(1, 1)
        ax.plot(self.returns.index, self.pnl[1:], 'b', label=self.name + ' (vol: %.2f, Sharpe: %.2f, MDD: %.2f)'%
                                                          (self.volatility(), self.sharpe_ratio(), self.max_dd()))
        ax.set_xlabel('Time')
        ax.set_ylabel('PnL')
        ax.legend()
        ax.tick_params(axis='x', rotation=45)
        ax.set_title(f'PnL of the strategy {self.name}')
        ax.grid(axis='x', linestyle='--')
        fig.tight_layout()
        fig.savefig(f'{self.name}.pdf')
        plt.show()
        
    def comparison_strategies(strategies):
        fig, ax = plt.subplots(1, 1)
        for strategy in strategies:
            ax.plot(strategy.returns.index, strategy.pnl[1:], label=strategy.name + ' (vol: %.2f, Sharpe: %.2f, MDD: %.2f)' %
                                                                            (strategy.volatility(), strategy.sharpe_ratio(), strategy.max_dd()))

        ax.set_xlabel('Time')
        ax.set_ylabel('PnL')
        ax.legend()
        ax.tick_params(axis='x', rotation=45)
        ax.set_title('PnL of the strategies')
        ax.grid(axis='x', linestyle='--')
        fig.tight_layout()
        plt.show()
```

```python
# On commence par la méthode la plus simple : l'EquallyWeighted portfolio
class EquallyWeighted(Strategy):

    def __init__(self, name, returns):
        weights_EW = np.full((len(returns), len(returns.columns)), 0.1, dtype=float) #On applique un poids de 1/10 à chaque stock
        returns_EW = (returns * weights_EW).sum(axis=1)
        super().__init__(name, returns_EW, returns)

ew_port = EquallyWeighted('EW_all', returns)
ew_port.illustrate()
ew_port.show_covariance_matrix() # On plot le matrice de covariance pour observer la corrélation entre les prix 
ew_port.show() #Affiche le portfolio
```

Markowitz Minimum Variance portfolio

Pour les deux prochaines stratégies, on aura besoin de la matrice de covariance sur l'année 2017-2018

```python
#On récupère les datas pour l'année 2017-2018

agg_price_mktcap_data_over_2017_2018 = agg_data.loc[(agg_data.index >= '31/12/2016') & (agg_data.index <= '31/12/2018')]
agg_price_mktcap_data_over_2017_2018_10_highest = agg_price_mktcap_data_over_2017_2018[highest_columns_10]
agg_price_data_over_2017_2018_10 = agg_price_mktcap_data_over_2017_2018_10_highest.T.filter(like = 'PX', axis=0).T
agg_price_data_over_2017_2018_10['MC FP Equity PX'].fillna(method = 'bfill', inplace = True)

#Calcul de la matrice de covariance 
returns_2017_2018 = agg_price_data_over_2017_2018_10.pct_change().dropna() 
emp_cov = EmpiricalCovariance().fit(returns_2017_2018)
cov = emp_cov.covariance_
```

```python
class markowitzMinimumVariance(Strategy):

    def __init__(self, name, returns,weightsOptimized):
        returns_MMV = (returns * weightsOptimized).sum(axis=1) # ici on utilise les weights optimisés par la fonction MPT ci-dessous 
        super().__init__(name, returns_MMV, returns)
```

```python
def weight_creator(df): # Pour l'étape initiale, on génère des random weights qu'on optimisera ensuite
        weights = abs(np.random.randn(len(df.columns)))
        weights = weights / np.sum(weights) # On normalise les weights
        return weights



def MPT(df, cov): # Generation des weights correspondant a la min variance 
    def returnsMPT(df, weights): 
        MeanReturns = df.mean()
        return np.dot(MeanReturns, weights)
    def portfolioRisk(weights, covMatrix): # On calcule le risk du portfolio, qui est proportionnel à la variance 
        return np.sqrt(np.dot(weights.T, np.dot(covMatrix, weights)))*np.sqrt(252)
    weights = []
    returns_MPT = []
    risks = []



    def optimizeWeightsMPT(df): 
        for i in range(10000):
            w = weight_creator(df)
            weights.append(w)
            returns_MPT.append(returnsMPT(df, w))
            risks.append(portfolioRisk(w, cov))
            
    optimizeWeightsMPT(df)
    Weights_minRisk = weights[risks.index(min(risks))]
    return Weights_minRisk
```

```python
mmv_port = markowitzMinimumVariance('MMV_all', returns, MPT(returns, cov))
mmv_port.illustrate()
mmv_port.show()
```

ERC Portfolio

Pour construire l'ERC portfolio, on va minimiser la fonction suivante :
$$ \sum_{i=1}^n \sum_{j=1}^n(w_i(Σw)_i - w_j(Σw)_j)^2 $$

Où :
- $ (w)_i $ est la i-ème composante du vecteur des poids.
- $ (Σw)_i $ est la i-ème ligne du vecteur résultant du produit de la matrice de covariance et du vecteur des poids.


Contraintes : 
- La somme des poids est égale à 1
- Les poids sont positifs

```python
class EqualRiskContribution(Strategy):

    def __init__(self, name, returns, weightsOptimized):
        returns_ERC = (returns * weightsOptimized).sum(axis=1)
        super().__init__(name, returns_ERC, returns)
```

```python
def ERC(df, cov) :
    def sqp(weights, covMatrix): # code la fonction Sequential Quadratic Programming 
        marginal_risk = covMatrix @ weights
        sqp_value = 0
        for i in range(len(weights)):
            for j in range(len(weights)):
                sqp_value += ((weights[i] * marginal_risk[i]) - (weights[j] * marginal_risk[j])) ** 2
        return sqp_value
    
    def min_risk_diff(weights, covMatrix): # On prend les weights qui minimise la variance de la risk contribution 
        constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
        bnds = len(weights) * [(0, 1),]
        result = minimize(lambda w: sqp(weights, covMatrix),
                   weights, bounds=bnds, constraints=constraints)
        return result.x
    
    weights = []
    sqp_v = []

    def optimizeWeightsERC(df): # Boucle pour générer les poids 
        for i in range(10000):
            w = min_risk_diff(weight_creator(df), cov)
            weights.append(w)
            sqp_v.append(sqp(w, cov))

    optimizeWeightsERC(df)
    Weights_minRisk = weights[sqp_v.index(min(sqp_v))]
    return Weights_minRisk
```

```python
erc_port = EqualRiskContribution('ERC_all', returns, ERC(returns, cov))
erc_port.illustrate()
erc_port.show()
```

On plot enfin les trois trajectoires du PnL avec un initial wealth de 1 million d'euros. On constate que la troisième stratégie présente le sharpe ratio le plus élevé. Il est toutefois important de noter que les weights des deux dernières stratégies présentent de l'aléatoire, et donc les résultats peuvent changer d'une exécution à l'autre.
